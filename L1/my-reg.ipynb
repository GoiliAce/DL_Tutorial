```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# data
n = 100
mu = 1.70
std = 0.03
x = norm.rvs(loc =mu,scale=std,size = (1,n)).reshape(-1,1)
y = (30*x + 2) +np.random.uniform(1,3,(1,n)).reshape(-1,1)
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
one_x = np.hstack((np.ones_like(x),x))
```

```python
w = np.array([0.,1.]).reshape([-1,1])
learning_rate = 0.001
epochs = 10000
costs = np.zeros((epochs,1))
for i in range(1,epochs):
    r = np.dot(one_x,w) - y
    costs[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate * np.sum(r)
    w[1] -= learning_rate *np.sum(np.multiply(r,x))
```

```python
x_line = np.linspace(1.62,1.80,5)
y_line = w[0] + w[1]*x_line
plt.plot(x,y,'o')
plt.plot(x_line,y_line)
plt.show()
```
